```
class NodoArbol:
  def __init__(self, value, left=None, rigth=None):
    self.data = value
    self.left =left
    self.rigth = rigth

class BinarySearchTree:
  def __init__(self):
    self.__root = None
  
  def insert(self, value):
    if self.__root == None:
      self.__root = NodoArbol(value,None,None)
    else:
      # preguntar si value es menor que root de ser el caso 
      #insertar ala isquierda PERO.. puede ser le caos que el
      # sub arbol tenga muchos elementos
      self.__insert_nodo__(self.__root, value)
  
  def __insert_nodo__(self, nodo, value):
    if nodo.data == value:
      pass # no se admiten valores repetidos
    elif value < nodo.data:# si es true va a la izq
      if nodo.left == None: # si hay espacio lo colaoca ahi
        nodo.left = NodoArbol(value)# inserta
      else:
        self.__insert_nodo__(nodo.left,value) # buscar en sub arbol izq
    else: #ir a la derecha
      if nodo.rigth == None:
        nodo.rigth = NodoArbol(value)
      else:
        self.__insert_nodo__(nodo.rigth, value) # buscar en sub arbol ider

  def buscar(self, value):
    if self.__root == None:
      return False
    else:
      # Haremos busqueda recursiva
      return self.__busca_nodo(self.__root, value)
  
  def __busca_nodo(self, nodo, value):
    if nodo == None:
      return None
    elif nodo.data == value:
      return nodo.data
    elif value < nodo.data:
      return self.__busca_nodo(nodo.left, value)
    else:
      return self.__busca_nodo(nodo.rigth, value)
  
  def transversal(self, format = "inorden"):
    if format == "inorden":
      self.__recorrido_in(self.__root)
    elif format == "preorden":
      self.__recorrido_pre(self.__root)
    elif format == "posorden":
      self.__recorrido_pos(self.__root)
    else:
      print("Formato de recorrido no valido")

  def __recorrido_in(self,nodo):
    if nodo != None:
      self.__recorrido_in(nodo.left)
      print(nodo.data, end=",")
      self.__recorrido_in(nodo.rigth)

  def __recorrido_pre(self, nodo):
    if nodo != None:
      print(nodo.data, end=",")
      self.__recorrido_pre(nodo.left)
      self.__recorrido_pre(nodo.rigth)
    
  def __recorrido_pos(self, nodo):
    if nodo != None:
      self.__recorrido_pos(nodo.left)
      self.__recorrido_pos(nodo.rigth)
      print(nodo.data, end=",")
  
  def remove(self, value):
    if self.__root == None:
      return False
    else:
      return self.__remove_rec(self.__root,value)
  
  def __remove_rec(self, nodo, value):
    if nodo == None:
      return None
    elif nodo.data == value:
      if nodo.rigth == None and nodo.left == None:
        #print("nodo a eliminar: ",nodo.data)
        return True 
      else:
        i=self.__busca_menor(nodo.rigth)
        #print("valor a cambiar:", i[0])
        if i[1] != 0:
          nodo.rigth=None
        #i[1]=0
        return i[0]

    elif value < nodo.data:
      x=self.__remove_rec(nodo.left, value)
      if x == True:
        nodo.left=None
      elif isinstance(x,int):
        nodo.left.data = x
      
    else:
      y=self.__remove_rec(nodo.rigth, value)
      #print("que hay: ",y)
      if y==True:
        nodo.rigth=None
      elif isinstance(y,int):
        #print("si entra".upper())
        nodo.rigth.data = y

  def __busca_menor(self,nodo):
    if nodo.left != None:
      x=self.__busca_menor(nodo.left)
      if x[1] != 0:
        nodo.left=None
        x[1]=0
      return x
    else:
      lista=[nodo.data,1]
      return lista
```
