```
def fnRec( x):
  if x==0:
    print("Stop")
  else:
    fnRec(x-1)
    print(x)

def main():
  print("inicio de programa")
  fnRec(5)
  print("fin del programa")
main()
```

```
def fibonacci(n):
  print("llamada:", n)
  if n == 1 or n == 0:
    return n
  else:
    return (fibonacci(n-1)+fibonacci(n-2))
print(fibonacci(20))
# n*(n-2) + 1
```
